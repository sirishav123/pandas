```python
plt.hist(df1['SAT'],color='red')
plt.xlabel('SAT')
plt.ylabel('Frequency')
plt.title('Distribution of SAT score')
plt.show()
```

```python
plt.hist(df1['SAT'], color='red')
plt.xlabel('SAT')
plt.ylabel('Frequency')
plt.title('Distribution of SAT Scores')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'ro')
plt.xlabel('SAT')
plt.ylabel('Accept')
plt.title('SAT vs Accept')
plt.show()
```

```python
# to create a scatter plot between the 'SAT' and 'Top10' columns using matplotlib.
plt.scatter(df1['SAT'], df1['Top10'], color='red', marker='o')
plt.xlabel('SAT')
plt.ylabel('Top10')
plt.title('SAT vs Top10')
plt.show()
```
